```python
# Task 2: Mark the success/failed launches for each site on the map
# Initialize the map centered at NASA JSC
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Loop through each launch in the dataset
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    launch_status = row['class']  # 1 for success, 0 for failure
    
    # Set marker color based on success or failure
    marker_color = 'green' if launch_status == 1 else 'red'
    
    # Add a marker with appropriate color
    folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=marker_color, icon='rocket')
    ).add_to(site_map)

# Display the map
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Define a function to determine marker color
def assign_marker_color(launch_outcome):
    return 'green' if launch_outcome == 1 else 'red'

# Apply the function to the 'class' column and create a new 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Display the first few rows to verify
spacex_df.head()

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Iterate over each row in spacex_df data frame
for index, record in spacex_df.iterrows():
    # Extract coordinates
    launch_coordinate = [record['Lat'], record['Long']]
    
    # Create a Marker object with appropriate icon color based on success or failure
    marker = folium.Marker(
        location=launch_coordinate,
        icon=folium.Icon(color='white', icon_color=record['marker_color']),  # Color is set dynamically
        popup=f"Launch Site: {record['Launch Site']}\nClass: {record['class']}"
    )
    
    # Add marker to the cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Coordinates of the launch site
launch_site_lat = 28.56342
launch_site_lon = -80.57674

# Coordinates for the closest features (Replace with actual values from MousePosition)
coastline_lat, coastline_lon = 28.56367, -80.57163
city_lat, city_lon = 28.3922, -80.6077
railway_lat, railway_lon = 28.57205, -80.58527
highway_lat, highway_lon = 28.5623, -80.5709

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    from math import sin, cos, sqrt, atan2, radians
    R = 6373.0  # Radius of the Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Compute distances
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Function to create markers with distances
def add_distance_marker(map_object, coord, distance, label):
    marker = folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{}</b></div>'.format(f"{label}: {distance:.2f} km"),
        )
    )
    map_object.add_child(marker)

# Add distance markers
add_distance_marker(site_map, [coastline_lat, coastline_lon], distance_coastline, "Coastline")
add_distance_marker(site_map, [city_lat, city_lon], distance_city, "City")
add_distance_marker(site_map, [railway_lat, railway_lon], distance_railway, "Railway")
add_distance_marker(site_map, [highway_lat, highway_lon], distance_highway, "Highway")

# Draw lines from launch site to features
folium.PolyLine([[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], color="blue", weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [city_lat, city_lon]], color="red", weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], color="green", weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], color="purple", weight=2).add_to(site_map)

# Display map
site_map
```
